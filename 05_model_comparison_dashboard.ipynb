# Model Comparison Dashboard

Compare the performance of Logistic Regression and Random Forest models using multiple evaluation metrics.

This notebook summarizes model performance and highlights the tradeoffs between interpretability and predictive power.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
comparison_df = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest"
    ],
    "Accuracy": [
        0.8078,
        0.8744
    ],
    "Default Recall": [
        0.17,
        0.62
    ],
    "Default Precision": [
        0.70,
        0.75
    ],
    "Default F1": [
        0.28,
        0.68
    ]
})

comparison_df
```

```python
comparison_df.plot(
    x="Model",
    y="Accuracy",
    kind="bar"
)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")

plt.show()
```

```python
comparison_df.plot(
    x="Model",
    y="Default Recall",
    kind="bar"
)

plt.title("Default Recall Comparison")
plt.ylabel("Recall")

plt.show()
```

## Findings

Random Forest outperformed Logistic Regression across all evaluation metrics.

While Logistic Regression provides a simple and interpretable baseline model, its ability to identify default borrowers was limited.

Random Forest achieved higher accuracy, recall, and F1-score, making it a more effective model for credit risk prediction.

Cross-validation and SHAP analysis further confirmed the robustness and interpretability of the Random Forest model.
